```python
import idx2numpy
mnist_x_train = idx2numpy.convert_from_file('/kaggle/input/mnistdataset/train-images.idx3-ubyte')
print(mnist_x_train.shape)
mnist_y_train = idx2numpy.convert_from_file('/kaggle/input/mnistdataset/train-labels.idx1-ubyte')
print(mnist_y_train.shape)
mnist_x_test = idx2numpy.convert_from_file('/kaggle/input/mnistdataset/t10k-images.idx3-ubyte')
print(mnist_x_test.shape)
mnist_y_test = idx2numpy.convert_from_file('/kaggle/input/mnistdataset/t10k-labels.idx1-ubyte')
print(mnist_y_test.shape)
```

```python
import numpy as np
train = np.genfromtxt('/kaggle/input/digit-recognizer/train.csv', delimiter = ',')
test = np.genfromtxt('/kaggle/input/digit-recognizer/test.csv', delimiter = ',')
```

```python
print(test[1:].shape)
```

```python
x_train = train[1:, 1:]
y_train = train[1:][:,0]
x_test = test[1:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
```

Training data: mnist_x_train, mnist_x_test, x_train

Corr. targets: mnist_y_train, mnist_y_test, y_train

To pred: **x_test**

```python
mnist_x_train = np.reshape(mnist_x_train, (-1, 28, 28, 1))
mnist_x_test = np.reshape(mnist_x_test, (-1, 28, 28, 1))
x_train = np.reshape(x_train, (-1, 28, 28, 1))
x_test = np.reshape(x_test, (-1, 28, 28, 1))
```

```python
print(mnist_x_train.shape)
print(mnist_x_test.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
mnist_x_train = mnist_x_train/255
mnist_x_test = mnist_x_test/255
x_train = x_train/255
x_test = x_test/255
```

```python
x_train.shape
```

```python
sample = 5
import matplotlib.pyplot as plt
img = mnist_x_train[sample]
print(img.shape)
plt.imshow(img)
print(mnist_y_train[sample])
```

```python
import tensorflow as tf

num_classes = 10
inputs = tf.keras.Input(shape=(28, 28, 1))

x = tf.keras.layers.Conv2D(filters = 32, kernel_size = (3, 3), padding = "same")(inputs)
x = tf.keras.layers.MaxPool2D(pool_size = (2, 2), strides = 2)(x)

x = tf.keras.layers.Conv2D(filters = 64, kernel_size = (3, 3), padding = "same")(x)
x = tf.keras.layers.MaxPool2D(pool_size = (2, 2), strides = 2)(x)

x = tf.keras.layers.Flatten()(x)

x = tf.keras.layers.Dense(32, activation = 'relu')(x)
x = tf.keras.layers.Dense(64, activation = 'relu')(x)
# x = tf.keras.layers.Dropout(rate = .2)(x)
# x = tf.keras.layers.Dense(, activation = 'relu')(x)

outputs = tf.keras.layers.Dense(num_classes, activation = 'softmax')(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)

# predictions = model(train_data)
```

```python
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.SGD(), loss = tf.keras.losses.SparseCategoricalCrossentropy(), metrics = ['accuracy'])
model.fit(mnist_x_train, mnist_y_train, validation_data = (mnist_x_test, mnist_y_test), epochs = 20)
```

```python
model.fit(x_train, y_train, epochs = 5)
```

```python
predictions = model.predict(x_test)
```

```python
print(predictions.shape)
```

```python
p = np.argmax(predictions, axis=1)
print(p.shape)
```

```python
p.shape[-1]
```

```python
import csv
count = p.shape[-1]
with open('/kaggle/working/submission.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['ImageId','Label'])
    for i in range(count):
        writer.writerow([i+1, p[i]])
```
